```python
# Import Dependencies
import pandas as pd
```

```python
assistance_csv = "Resources/Assistance Received by Country.csv"
emission_csv = "Resources/CO2 Emission by Country.csv"
employment_csv = "Resources/Employment by Country.csv"
gdp_csv = "Resources/GDP by Country.csv"
```

```python
assistance_df = pd.read_csv(assistance_csv)
emission_df = pd.read_csv(emission_csv)
```

```python
assistance_df.head()
```

```python
emission_df.head()
```

```python
# Merge the two DataFrames together based on the country they share
merge1_df = pd.merge(assistance_df, emission_df, on="Country")
merge1_df.head()
```

```python
employment_df = pd.read_csv(employment_csv)
gdp_df = pd.read_csv(gdp_csv)
```

```python
employment_df.head()
```

```python
gdp_df.head()
```

```python
# Merge the two DataFrames together based on the country they share
merge2_df = pd.merge(employment_df, gdp_df, on="Country")
merge2_df.head()
```

```python
# Merge the two merged DataFrames together based on the country they share
merge3_df = pd.merge(merge1_df, merge2_df, on="Country")
merge3_df.head()
```

```python
merge3_df.to_csv("world_data.csv", encoding = "utf-8") 
```
