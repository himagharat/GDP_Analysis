# Effect of Financial Aid on Sustaintability Indicators

<br>

This project aims to explore if there is some significant effect of financial aid given to some countries in their Sustaintability Indexes, such as Carbon dioxide average emissions, employment rate and gross domestic product.

```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

warnings.filterwarnings('ignore')
RANDOM_SEED = 42
```

## Collecting the Data

```python
# reading dataset for countries with no financial aid
c_0 = pd.read_csv('countries_0.csv')
c_0.head(1)
```

```python
# reading dataset for countries with financial aid
c_1 = pd.read_csv('countries_assistance.csv')
c_1.head(1)
```

```python
# joining datasets
data = pd.concat([c_0, c_1], axis = 0)

# checking shapes
print('Not Aided Countries data: ', c_0.shape)
print('Aided Countries data: ', c_1.shape)
print('Full data: ', data.shape)
```

## Data cleaning

```python
# checking for missing data
data.isnull().sum()
```

```python
# checking data type
data.info()
```

```python
# removing unused columns
data = data.drop(['Unnamed: 0'], axis = 1)
data.head(1)
```

```python
# creating categorical column indicating the presence of financial aid
data['Aided'] = 'Not aided'
data.loc[
    data['Mean Assistance (2011-2016)'] > 0,
    'Aided'
] = 'Aided'

# checking relative distribution of aided countries
data['Aided'].value_counts(normalize = True)
```

## Exploratory Data Analysis

<br>

### Histogram and Distribution

```python
# defining function for doing some distribution and boxplot analysis
def hist_box_plot(data, var_name):
    
    fig = plt.figure(figsize = (20,10))
    
    # plot histogram by aid class
    ax1 = fig.add_subplot(1,2,1)
    ax1.hist(data.loc[
        data['Aided'] == 'Not aided',
        'Mean CO2 (2011-2016)'
    ], label = 'Not aided countries', color = 'tomato', alpha = 0.5,
            density = True)
    ax1.hist(data.loc[
        data['Aided'] == 'Aided',
        'Mean CO2 (2011-2016)'
    ], label = 'Aidad countries', color = 'green', alpha = 0.5,
            density = True)
    ax1.set_xlabel(var_name, size = 14)
    ax1.set_ylabel('Density', size = 14)
    ax1.set_title('Distribution of ' + var_name, size = 18)
    ax1.legend()
    ax1.grid()    
        
    # plot boxplot by aid class
    ax2 = fig.add_subplot(1,2,2)
    sns.boxplot(data['Aided'], data[var_name], hue = data['Aided'])
    ax2.set_ylabel(var_name, size = 14)
    ax2.set_xlabel('')
    ax2.set_title('Boxplot of ' + var_name, size = 18)
    ax2.legend()
    ax2.grid()
```

```python
# EDA for CO2
hist_box_plot(data, 'Mean CO2 (2011-2016)')
```

```python
# EDA for Employment
hist_box_plot(data, 'Mean Employment (2011-2016)')
```

```python
# EDA for Mean GDP
hist_box_plot(data, 'Mean GDP (2011-2016)')
```

### Scatterplot and Correlation Analysis (only for Aid > 0)

```python
def scatter_analysis(data, var_name, x_var = 'Mean Assistance (2011-2016)'):
    
    # select only data with aid > 0
    df = data.loc[data['Aided'] == 'Aided']
    
    # plot graph
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(1,1,1)
    ax.plot(df[x_var], df[var_name], 'ko')
    ax.set_ylabel(var_name, size = 14)
    ax.set_xlabel(x_var, size = 14)
    ax.set_title('Relationship between %s and %s'%(x_var, var_name), size = 18)
    ax.grid()
```

```python
# scatteplot for mean CO2
scatter_analysis(data, 'Mean CO2 (2011-2016)')
```

```python
# scatteplot for mean Employment
scatter_analysis(data, 'Mean Employment (2011-2016)')
```

```python
# scatteplot for mean GDP
scatter_analysis(data, 'Mean GDP (2011-2016)')
```

```python
# data for correlation
data_corr = data.loc[
    data['Aided'] == 'Aided',
    ['Mean Assistance (2011-2016)', 'Mean CO2 (2011-2016)', 
     'Mean Employment (2011-2016)', 'Mean GDP (2011-2016)']
]

# calculating the Spearman's correlation coefficient
plt.figure(figsize = (10,10))
sns.heatmap(data_corr.corr(method = 'spearman'),
           vmin = -1, vmax = 1,
           cmap = 'coolwarm',
           annot = True)
```

### Feature Engineering

```python
# creating new features by applying the log function
data['logAssist'] = np.log(data['Mean Assistance (2011-2016)'])
data['logCO2'] = np.log(data['Mean CO2 (2011-2016)'])
data['logEmploy'] = np.log(data['Mean Employment (2011-2016)'])
data['logGDP'] = np.log(data['Mean GDP (2011-2016)'])

# checking new features
data.head(1)
```

```python
# scatterplot - logCO2
scatter_analysis(data, 'logCO2', 'logAssist')
```

```python
# scatterplot - logEmploy
scatter_analysis(data, 'logEmploy', 'logAssist')
```

```python
# scatterplot - logGDP
scatter_analysis(data, 'logGDP', 'logAssist')
```

```python
# cleaning outliers from dataset of CO2
dataCO2 = data.loc[
    (data['logCO2'] > 3) &
    (data['logAssist'] > 15)
]

# repeating scatter analysis
scatter_analysis(dataCO2, 'logCO2', 'logAssist')
```

```python
# cleaning outliers from dataset of Employment
dataEmploy = data.loc[
    (data['logEmploy'] > 1.5) &
    (data['logAssist'] > 16)
]

# repeating scatter analysis
scatter_analysis(dataEmploy, 'logEmploy', 'logAssist')
```

```python
# cleaning outliers from dataset of Employment
dataGDP = data.loc[
    (data['logGDP'] > 6) &
    (data['logAssist'] > 0)]

# repeating scatter analysis
scatter_analysis(dataGDP, 'logGDP', 'logAssist')
```

#### Model Training - CO2

```python
# extracting predictors and targets
x_CO2 = dataCO2['logAssist'].values
y_CO2 = dataCO2['logCO2'].values
```

```python
scaler = StandardScaler()
scaler = scaler.fit(x_CO2.reshape(-1,1))
x_CO2 = scaler.transform(x_CO2.reshape(-1,1))
modelCO2 = LinearRegression()
modelCO2.fit(x_CO2,y_CO2)
```

```python
dataCO2['CO2 prediction']= modelCO2.predict(x_CO2)
```

```python
dataCO2
```

```python
dataCO2.to_csv("dataCO2.csv", encoding='utf-8')
```

### Model Training - Employment Rate

```python
# extracting predictors and targets
x_Emp = dataEmploy['logAssist'].values
y_Emp = dataEmploy['logEmploy'].values
```

```python
scaler = StandardScaler()
scaler = scaler.fit(x_Emp.reshape(-1,1))
x_Emp = scaler.transform(x_Emp.reshape(-1,1))
modelEmp = LinearRegression()
modelEmp.fit(x_Emp,y_Emp)
```

```python
dataEmploy['Employment prediction']= modelEmp.predict(x_Emp)
```

```python
dataEmploy
```

```python
dataEmploy.to_csv("dataEmploy.csv", encoding='utf-8')
```

### Model Training - GDP

```python
# extracting predictors and targets
x_GDP = dataGDP['logAssist'].values
y_GDP = dataGDP['logGDP'].values
```

```python
scaler = StandardScaler()
scaler = scaler.fit(x_GDP.reshape(-1,1))
x_GDP = scaler.transform(x_GDP.reshape(-1,1))
modelGDP = LinearRegression()
modelGDP.fit(x_GDP,y_GDP)
```

```python
dataGDP['GDP prediction']= modelGDP.predict(x_GDP)
```

```python
dataGDP
```

```python
dataGDP.to_csv("dataGDP.csv", encoding='utf-8')
```
