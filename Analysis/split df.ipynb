```python
import pandas as pd
```

```python
world1_path=r'..\23-Final-Project\world_data1.csv'
```

```python
world1 = pd.read_csv(world1_path, low_memory=False, encoding="utf-8")
world1
```

```python
world1_df = pd.DataFrame(world1, columns = ['Country', 'Mean Assistance (2011-2016)', 'Mean CO2 (2011-2016)', 'Mean Employment (2011-2016)', 'Mean GDP (2011-2016)']) 
                  
# splitting the dataframe into 2 parts 
# on basis of 'Mean Assistance (2011-2016)' column values 
# using Relational operator 
world2_df = world1_df[world1_df['Mean Assistance (2011-2016)'] == 0 ] 
  
# printing world2_df 
world2_df
```

```python
world2_df.to_csv('countries_0.csv', encoding='utf-8')
```

```python
world3_df = world1_df[world1_df['Mean Assistance (2011-2016)'] >= 1 ] 
world3_df
```
