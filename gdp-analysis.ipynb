```python
# Import Dependencies
import pandas as pd
```

```python
assistance_csv = "Resources/Assistance Received by Country.csv"
emission_csv = "Resources/CO2 Emission by Country.csv"
employment_csv = "Resources/Employment by Country.csv"
gdp_csv = "Resources/GDP by Country.csv"
```

```python
assistance_df = pd.read_csv(assistance_csv)
emission_df = pd.read_csv(emission_csv)
employment_df = pd.read_csv(employment_csv)
gdp_df = pd.read_csv(gdp_csv)
```

```python
# Merge the four DataFrames together based on the country they share
merged_df = pd.merge(assistance_df, emission_df, employment_df, gdp_df, on="unnamed")
merged_df.head()
```

```python
# Rename columns so that they are differentiated
merged_df = merged_df.rename(columns={"Open_x": "Bitcoin Open", "High_x": "Bitcoin High", "Low_x": "Bitcoin Low",
                                      "Close_x": "Bitcoin Close", "Volume_x": "Bitcoin Volume", "Market Cap_x": "Bitcoin Market Cap"})

merged_df = merged_df.rename(columns={"Open_y": "Dash Open", "High_y": "Dash High", "Low_y": "Dash Low",
                                      "Close_y": "Dash Close", "Volume_y": "Dash Volume", "Market Cap_y": "Dash Market Cap"})

merged_df.head()
```
